Create NumPy arrays in multiple ways and inspect their shape and dtype.

```python
import numpy as np
a = np.array([5, 3, 7, 20])
b = np.zeros([5,2])
o = np.ones(3)
arr = np.arange(0, 9, 3)
l = np.linspace(4, 10, 3)
ran = np.random.rand(3, 2)

print(f"linsapace: {arr}")
print(f"linsapace: {l}")
print(f"linsapace: {ran}")
```

Create a 2D array of shape (4, 4) filled with values 1-16 and print its shape and dtype.

```python
array_2d = np.arange(1, 17).reshape(4, 4)

print(array_2d)
```

Use slicing to extract the second column and the last row.

```python
second_column = array_2d[:, 1]
last_row = array_2d[-1, :]

print("second_column= ", second_column)
print("\nlast_row= ", last_row)
```

Use a boolean mask to select all values greater than the array's mean.

```python
mean = array_2d.mean()
mask = array_2d > mean

value_above_mean = array_2d[mask]

print(f"Array Mean: {mean}")
print(f"\nBoolean Mask: {mask}")
print(f"Value above mean: {value_above_mean}")
```

Add a 1D row array to every row of a 2D array using broadcasting, and verify the result manually.

```python
row_1d = np.array([3, 6, 9, 11])

broadcast = array_2d + row_1d

print(f"Original 2D Array:{array_2d}")
print(f"\n1D Row: {row_1d}")
print(f"\nBroadcast Result: {broadcast}")
```
