A funtion that take a list of numbers and returns the mean, min and max as a dictionary

```python
def operationFunc(numbers):
    mean_val = sum(numbers) / len(numbers)
    min_val = min(numbers)
    max_val = max(numbers)

    return {
        "mean": mean_val,
        "min": min_val,
        "max": max_val,
    }
```

```python
#Test
sample = [10,25,17,4]
print(operationFunc(sample))
```

For loop that filters even numbers as a single list comprehension

```python
numbers = [1,3,2,5,23,52,14,8]
evens = []
for num in numbers:
    if (num % 2 == 0):
        evens.append(num)
print(evens)
```

Object-Oriented Programming (OOP) Data Record Class

```python
class DataRecord:
    def __init__(self, id, value):
        self.id = id
        self.value = value
    def is_above_value(self, aboveValue = 10):
        return self.value >= aboveValue

record = DataRecord(id = 1, value = 5)
print(f"Is above? {record.is_above_value()}")
```
