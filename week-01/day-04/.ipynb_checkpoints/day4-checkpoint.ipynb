 Load a provided CSV dataset into a DataFrame and report its shape, columns, and dtypes

```python
import pandas as pd

df = pd.read_csv("tested.csv")
print(f"first 5 rows: {df.head()}")
print(f"(rows, columns): {df.shape}")
print(f"column names, non-null counts, dtype: {df.info()}")
print(f"summary statistics for numeric columns: {df.describe()}")
```

 Count missing values per column and handle them (fill or drop) with a documented justification.

```python
print("Counting of missing values:")
print(df.isnull().sum())

#Filling the missing age values with mean
mean_age = df["Age"].mean()
df["Age"] = df["Age"].fillna(mean_age) 

#Drop rows with any missing value
df.dropna(inplace=True) 
```

Justification: Age Column: Filled missing gaps with the average age to keep the dataset size intact without altering the overall average.
Other Columns: Deleted remaining rows with missing parts to ensure the final data is completely clean and ready for analysis.

Filter the data to a meaningful subset (e.g. rows above a threshold) and describe what you find.

```python
# Filter for passengers who paid a Fare above the average price
fare_threshold = df['Fare'].median() #average
high_fare_df = df[df['Fare']>fare_threshold]

print(f"Original Passengers: {len(df)}")
print(f"High-Fare Passengers: {len(high_fare_df)}")
```

What i found: half of the passengers (45 out of 91) paid a higher-than-average ticket price.

Use groupby to compute an aggregate statistic per category and interpret the result in a Markdown cell.

```python
servival_by_sex = df.groupby('Sex')['Survived'].mean()

print(f"Survival Rate by Gender: {servival_by_sex}")
```

The Result: all females in Female Passengers subset survived and no male survived in Male Passengers subset
