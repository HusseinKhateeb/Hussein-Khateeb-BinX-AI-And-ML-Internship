Load and clean a provided dataset with Pandas (handling any missing values).

```python
import pandas as pd

df = pd.read_csv("tested.csv")

#Filling the missing age values with mean
mean_age = df["Age"].mean()
df["Age"] = df["Age"].fillna(mean_age) 

#Drop rows with any missing value
df.dropna(inplace=True) 
```

Use NumPy to compute at least one derived numeric feature or summary statistic.

```python
import numpy as np

age_mean = np.mean(df['Age'])
age_std = np.std(df['Age'])

df['Age_Score'] = (df['Age'] - age_mean) / age_std

print("DataFrame with Derived Features:")
print(df[['Age', 'Age_Score']].head())
```

The negative z-scores for younger ages (like 22 and 23) show they are below the group average, while 47 and 48 sit around 0.5–0.6 standard deviations above it.

Produce at least three labeled plots (including a histogram and a scatter plot) exploring the data.

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid')
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

#Plot 1
sns.histplot(df['Age'], color='skyblue', ax=axes[0])
axes[0].set_title('Age Distribution')
axes[0].set_xlabel('Age')
axes[0].set_ylabel('Frequency')

#Plot 2 
axes[1].scatter(df['Age'], df['Fare'], color='red')
axes[1].set_title('Age Vs Fare')
axes[1].set_xlabel('Age')
axes[1].set_ylabel('Fare')

#Plot 3
axes[2].boxplot(df['Age'].dropna())
axes[2].set_title('Age Outlier Analysis')
axes[2].set_ylabel('Age')
axes[2].set_xlabel([])

plt.tight_layout()
plt.show()
```

Write Markdown cells explaining what each visualization reveals about the data.

Histogram (Age Distribution): Shows most passengers in this dataset are concentrated between 20 and 50 years old.
Scatter Plot (Age vs. Fare): Shows that most ticket prices are clustered under $100, though a few individuals paid much higher fares (up to $500+).
Box Plot (Age Outliers): Shows the median age is around 37, with the middle 50% of ages sitting between ~27 and ~50.
